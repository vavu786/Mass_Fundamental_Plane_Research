##### I want to fit this data with a function motivated by the literature but need to work it out in log space

Rewriting the MCMC code in log space:

The equation for the single power law is quite easily rewritten in log space:

$y = \alpha \Big( \frac{x}{5*10^{10}}\Big) ^{\beta}$ simply becomes:


$\log_{10}(y) = \log_{10}\Big[\alpha \Big( \frac{x}{5*10^{10}}\Big) ^{\beta}\Big]$

$ \log_{10} y =  \log_{10} (\alpha) + \beta \log_{10} \Big( \frac{x}{5*10^{10}}\Big) $

$ \log_{10} y =  \log_{10} (\alpha) +\beta \log_{10}( x) - \beta \log_{10}\Big(5*10^{10}\Big)$



$ \log_{10} y =  A +B \log_{10}( x) - B\log_{10}\Big(5*10^{10}\Big)$

Need to fit A and B

#### 1st: We need to define the following for emcee

```python
# I use the standard definition of log likelihood
# sigma is the covariance between the x & y uncertainties
def log_likelihood(theta, x, y, xerr, yerr):
    b, a = theta
    v = np.array([-b, 1])
    model = a + b*(x) - b*np.log10(5e10)  #the form that is being fit 
    # the likelihood is sum of the lot of normal distributions
    #S=np.cov(yerr, xerr) ### S is the covariance of xerr & yerr but need it for each data point
    Sigma2 = np.dot(np.dot(S, v), v)
    return -0.5 * np.sum((model-y)**2 / Sigma2 + np.log(Sigma2))

# Assume uniform priors over the range I let a and b vary
def log_prior(theta):
    b, a = theta
    if 0 < b < 1 and 0.1 < a < 2:
        return 0
    return -np.inf

def log_probability(theta, x, y, xerr, yerr):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, xerr, yerr)
```

#### MCMC!!!!!!!

```python
pos = np.array([0.3, 0.5]) + 1e-1 * np.random.randn(50, 2)
# Initialize 50 walkers within a gaussian sphere centered on beta=0.3, alpha=0.8

nwalkers, ndim = pos.shape

# As far as I know, MCMC needs symmetric uncertainties... so I take the larger of the 2 (i.e., max[upper, lower])
xerrors = np.nanmax(np.vstack((zbin1['lowerM_unc'], zbin1['upperM_unc'])),axis=0)
yerrors = np.nanmax(np.vstack((zbin1['lowerR_unc'], zbin1['upperR_unc'])),axis=0)

# then the covariance matrix:
N= len(zbin1['sizes'])
S = np.zeros((N, 2, 2))
for n in range(N):
    S[n] = np.cov(yerrors, xerrors)


# HERE I only run 1000 interations so it runs quickly (way too little! should be increased to probably ~5000)
sampler_SF_zbin1_allM = emcee.EnsembleSampler(nwalkers, ndim, log_probability, 
                                              args=(zbin1['masses'], zbin1['sizes'],xerrors,yerrors))
                                              
sampler_SF_zbin1_allM.run_mcmc(pos, 1000, progress=True)

```

```python
fig, axes = plt.subplots(2, figsize=(10, 7), sharex=True)
samples = sampler_SF_zbin1_allM.get_chain()
labels = ["alpha", "beta"]
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

```python
def model(bottom, top, mod): return np.log10(10**mod[1]* (np.logspace(bottom,top)/(5e10))**mod[0]) 
def sample_walkers(bottom, top, sigma_level, nsamples,flattened_chain):
    models = []
    draw = np.floor(np.random.uniform(0,len(flattened_chain),size=nsamples)).astype(int)
    thetas = flattened_chain[draw]
    for i in thetas:
        mod = model(bottom, top, i)
        models.append(mod)
    spread = sigma_level*np.std(models,axis=0)
    med_model = np.median(models,axis=0)
    return med_model,spread

```
